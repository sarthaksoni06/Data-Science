# EDA

```python
sns.pairplot(df,hue='Sub-Category')
```

```python
sns.countplot(df['Ship Mode'])
```

```python
#we can observe that,Standard Class is the highest in ship modes, and same day is lowest prefered shipment mode 
sns.countplot(df['Segment'])
```

```python
#We can observe that Cosumer in highest in Segment and home office being the least 
sns.countplot(df['Region'])
```

```python
# we can observe that Western region is best performimg , and we can see that southern region is the lowest performing 

plt.xticks(rotation=90)
sns.countplot(df['Category'])
```

```python
#we can observe that the office suppies is in demand while the other two categories are at par with each other 
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.countplot(df['Sub-Category'])
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.countplot(df['State'])
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.barplot('State','Profit',data=df)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.barplot('Sub-Category','Profit',data=df)
```
